```python
X = df.drop(columns = ['salary_in_usd'])
y = df['salary_in_usd']
```

```python
X = pd.get_dummies(X)
```

```python
X # galat aa rha hai true false aana chahiye
```

```python
# Train - Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
# Linear Regression
# KNNR
# DTR   
# RFR
```

```python
from sklearn.preprocessing import StandardScaler
se = StandardScaler()
X_train = se.fit_transform(X_train)
X_test_scaled = se.fit_transform(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, root_mean_squared_error
def evaluate_model(model, X_train, y_train, X_test, y_test): # for model-> LinearR, KNN, DT, RF
    train_pred = model.predict(X_train)
    test_pred = model.predict(X_test)

    print(f"Accuracy for your model {model} is :")
    
    #R2 Score
    print(f"Training R2 Score for {model} is : {r2_score(y_train, train_pred)}")
    print(f"Testing R2 Score for {model} is : {r2_score(y_test, test_pred)}")
    
    #Mean Absolute Error(MAE)
    print(f"Training Mean Absoulute Error for {model} is : {mean_absolute_error(y_train, train_pred)}")
    print(f"Testing Mean Absoulute Error for {model} is : {mean_absolute_error(y_test, test_pred)}")
    
    #Root Mean Squared Error(RMSE)
    print(f"Training Root Mean Squared Error for {model} is : {root_mean_squared_error(y_train, train_pred)}")
    print(f"Testing Root Mean Squared Error for {model} is : {root_mean_squared_error(y_test, test_pred)}")
```

```python
from sklearn.linear_model import LinearRegression
linear_regression_model = LinearRegression()
```

```python
linear_regression_model.fit(X_train, y_train)
```

```python
evaluate_model(linear_regression_model,X_train, y_train, X_test, y_test)
```

```python
y_pred_test = linear_regression_model.predict(X_test)
```

```python
y_pred_train = linear_regression_model.predict(X_train)
```

```python
from sklearn.metrics import r2_score
# r2_score(y_test, y_pred_test)
r2_score(y_train, y_pred_train)

# MODEL IS OVERFIT : because it is not working good enough in testing
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge
```

```python
ridge = Ridge()
```

```python
ridge_prem = {
    "alpha" : [0.1, 1, 10, 50, 1000]
}
```

```python
ridge_search = GridSearchCV(ridge, ridge_prem, cv = 5)
```

```python
ridge_search.fit(X_train, y_train)
```

```python
best_ridge = ridge_search.best_estimator_
```

```python
evaluate_model(best_ridge, X_train, y_train, X_test, y_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn_model = KNeighborsRegressor()
```

```python
knn_param = {
    'n_neighbors' : [1, 3, 5, 7, 9, 11]
}
```

```python
knn_search = GridSearchCV(knn_model, knn_param, cv = 5)
```

```python
knn_search.fit(X_train, y_train)
```

```python
best_knn = knn_search.best_estimator_
```

```python
evaluate_model(best_knn, X_train, y_train, X_test, y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr_model = DecisionTreeRegressor()
```

```python
dtr_param = {
    'max_depth' : [3, 5, 7, 10],
    'min_samples_split' : [2, 5, 10]
}
```

```python
tree_search = GridSearchCV(dtr_model, dtr_param, cv = 5)
```

```python
tree_search.fit(X_train, y_train)
```

```python
best_tree = tree_search.best_estimator_
```

```python
evaluate_model(best_tree, X_train, y_train, X_test, y_test)
```
